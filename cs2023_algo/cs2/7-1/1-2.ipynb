整数のリストを引数とする以下の関数 f(lst)について、
リストの長さを 100000 から 1000000 まで 100000 刻みで増やしたとき、
それぞれの長さのリストに対する実行時間を計測するプログラムを jupyter 上で作成しなさい。
ただし、リストは、与えられた長さを n としたとき、0 以上 2n 未満の整数からランダムに n 個の要素を
サンプリング(random.sample 関数を用いること)したものを作成すること。


```python
def f(lst):
    result = 0
    i = len(lst)
    while i > 0:
        result += lst[i - 1]
        i = i // 2
    return result


import random
import matplotlib.pyplot as plt


def make_list(n):
    lst = random.sample(range(2 * n), n)
    f(lst)


for i in range(1, 11):
     %timeit make_list(100000 * i)

y = [46.9, 110, 195, 263, 368, 409, 487, 605, 651, 709]

x = []
for i in range(1, 11):
    x.append(100000 * i)

plt.plot(x, y)
plt.show()
```
