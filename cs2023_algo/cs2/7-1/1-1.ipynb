100000 円を、年利 2%で運用することを考える。単利で運用した場合と、複利で運用した場合について、それぞれの 50 年間の利子を含む残金を、一つの図に重ねて描画するプログラムを作成しなさい。ただし、以下の条件を満たすようにすること。

凡例は単利を”Simple”とし、複利を”Compound”とする。
キャプションは”Simple and Compound interest”とすること。
横軸のラベルは”Year”、縦軸のラベルは”Amount”とすること。
単利のための線は実線とすること。
複利のための線は実線とすること。


```python
import matplotlib.pyplot as plt

years = [x for x in range(1, 51)]
simple = [100000 + 100000 * 0.02 * x for x in years]
compound = [100000 * (1 + 0.02) ** x for x in years]

plt.plot(years, simple, label="Simple")
plt.plot(years, compound, label="Compound")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Amount")
plt.title("Simple and Compound interest")
plt.show()
```
