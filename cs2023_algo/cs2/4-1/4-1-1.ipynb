```python
def find_center(lst):
    center = len(lst) // 2
    return lst.index(center)
```

```python
import matplotlib.pyplot as plt


for i in range(1, 11):
    target = list(range(i * 100000))
    %timeit find_center(target)
```

```python
xs = [x * 100000 for x in range(1, 11)]
ys = [0.538, 0.964, 1.22, 1.64, 1.97, 2.38, 2.76, 3.23, 3.68, 3.93]

print(xs)
plt.plot(xs, ys, "o")
plt.xlabel("Length")
plt.ylabel("Time")
plt.title("Time of find_center")
plt.show()
```
