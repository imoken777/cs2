## 課題2-2-1
x 軸を n 、y 軸を計測した時間として、案1と案2の計測時間を、それぞれ描画してみましよう

```python
import matplotlib.pyplot as plt
import time
```

```python
def fib1(n):
    if n <= 0:
        return 0
    if n == 1:
        return 1
    return fib1(n-1) + fib1(n-2)

def fib2(n):
    lst = [0] * (n + 1)
    lst[0] = 0
    lst[1] = 1
    for i in range(2, n + 1):
        lst[i] = lst[i - 1] + lst[i - 2]
    return lst[n]
```

```python
time1 = []
n1 = []

for i in range(5, 41, 5):
    start = time.perf_counter()
    fib1(i)
    time1.append(time.perf_counter() - start)
    n1.append(i)


```

```python
time2 = []
n2 = []

for i in range(5, 41, 5):
    start = time.perf_counter()
    for _ in range(1000):
        fib2(i)
    end = time.perf_counter()
    time2.append((end- start) / 1000)
    n2.append(i)
```

```python
plt.plot(n1, time1,marker = 'o', linestyle = 'none' ,label = "fib1" )
plt.title("Time of fib1(n)")
plt.xlabel("n")
plt.ylabel("Time(s)")
plt.legend()
plt.show()
```

```python
plt.plot(n2, time2,marker = 'o',linestyle = 'none' ,label = "fib2")
plt.title("Time of fib2(n)")
plt.xlabel("n")
plt.ylabel("Time(s)")
plt.legend()
plt.show()
```
