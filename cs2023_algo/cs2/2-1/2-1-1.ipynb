## 課題 2-1-1

x 軸をリストの長さ 、y 軸を計測した時間として 、 線形探索とニ分探索を
10000 回繰り返した結果の時間をプロットしてみましよう


```python
import matplotlib.pyplot as plt
import time
```

```python
def liner_search(lst, x):
    for item in lst:
        if item == x:
            return True
    return False
```

```python
def binary_search(lst, x):
    start = 0
    end = len(lst)
    while start < end:
        center = start + (end - start) // 2
        if lst[center] == x:
            return True
        elif lst[center] > x:
            end = center
        else:
            start = center + 1
    return False
```

```python
liner_times = []
binary_times = []
list_lengths = range(1000, 11000, 1000)
```

```python
for n in list_lengths:
    lst = list(range(n))
    target = n // 2

    start = time.time()
    for _ in range(10000):
        liner_search(lst, target)
    end = time.time()
    liner_times.append((end - start) * 1000)

    start = time.time()
    for _ in range(10000):
        binary_search(lst, target)
    end = time.time()
    binary_times.append((end - start) * 1000)
```

```python
plt.plot(list_lengths, liner_times, label="Liner search")
plt.plot(list_lengths, binary_times, label="Binary search")
plt.xlabel("Length")
plt.ylabel("Time(ms)")
plt.xlim(0, 10000)
plt.ylim(0, 2500)
plt.title("Liner search and Binary search")
plt.legend()
plt.show()
```
