## 1-1-2 Matplotlib でグラフを描画する

```python
import matplotlib.pyplot as plt
import math
```

```python
xs = [0,1,2,3]
ys = [0,3,2,3]
ys2 = [3,2,1,0]

plt.plot(xs, ys, 'o',label='sample')
plt.plot(xs, ys2, 'x',label='sample2')

plt.title('Sample graph')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

## 課題1-1-2
10000 円を、金利3%で運用するとします 。単利で運用した場合と、複利で運用した場合について、それぞれの20年間の利子を含む残金を、以下のように重ねて描画しなさい 。

```python
xs  = [0]
ys = [10000]

for i in range(1, 21):
    xs.append(i)
    ys.append((ys[0] * 0.03) + ys[i-1])

xs2 = [0]
ys2 = [10000]

for i in range(1, 21):
    xs2.append(i)
    ys2.append((ys[i-1] * 0.03) + ys2[i-1])

plt.plot(xs, ys,label='Simple')
plt.plot(xs2, ys2,label='Compound')

plt.title('Simple and Compound interest')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()

plt.show()
```
