## 課題3-1-2
選択ソートを実装し、sortedの場合と同様の手順で、計算時間をプロットしてください
ここでは、リストの長さを100から1000まで1OO刻みで増やして

```python
import random
import matplotlib.pyplot as plt
```

```python
def selection_sort(lst):
    length = len(lst)
    for i in range(length):
        min_index = i
        for j in range(i + 1, length):
            if lst[j] < lst[min_index]:
                min_index = j
        if i != min_index:
            tmp = lst[i]
            lst[i] = lst[min_index]
            lst[min_index] = tmp
    return lst
```

```python
for i in range(1, 11):
    target = random.sample(range(1000000), i * 100)
    %timeit selection_sort(target)
```

```python
xs = [i for i in range(100, 1001, 100)]
ys = [0.248, 0.944, 1.7, 3.1, 5.04, 7.74, 10.3, 13.9, 17.7, 20.5]

plt.plot(xs, ys, marker = 'o', linestyle = 'none')
plt.title('Time of selection_sort(n)')
plt.xlabel('n')
plt.ylabel('Time(ms)')
plt.show()
```
