## 課題１
以下の５つの関数について、$1 \leqq x \leqq 8$の区間のグラフを重ねて描画する

- $ y = x $ 
- $ y = x^2 $
- $ y = 2^x $
- $ y = \log x $
- $ y = x \log x $




```python
import matplotlib.pyplot as plt
import math

step = 0.1
xs = [i * step for i in range(int(1/step), int(8/step) + 1)]

ys1 = [i for i in xs]
ys2 = [i ** 2 for i in xs]
ys3 = [2 ** i for i in xs]
ys4 = [math.log(i) for i in xs]
ys5 = [i * math.log(i) for i in xs]

plt.plot(xs, ys1, label='y=x')
plt.plot(xs, ys2, label='y=x^2')
plt.plot(xs, ys3, label='y=2^x')
plt.plot(xs, ys4, label='y=log(x)')
plt.plot(xs, ys5, label='y=xlog(x)')

plt.xlim(1, 10)
plt.xticks(range(1, 11))
plt.legend()

plt.show()
```

## 発展課題
関数$y = \frac{1}{x} $について$-10 \leqq x \leqq 10$の区間のグラフを描画する Code セルを作成してください

```python
import matplotlib.pyplot as plt


xs_pos = [i * 0.1 for i in range(1, 101)]
xs_neg = [-i * 0.1 for i in range(1, 101)]

ys_pos = [1 / x for x in xs_pos]
ys_neg = [1 / x for x in xs_neg]

plt.plot(xs_pos, ys_pos, label='y=1/x', color='blue')
plt.plot(xs_neg, ys_neg, color='blue')

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = 1/x')
plt.legend()

plt.show()
```
