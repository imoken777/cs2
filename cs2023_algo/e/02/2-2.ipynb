```python
import matplotlib.pyplot as plt
import csv

month = []
temperature = []
with open("weather_tokyo_2020.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        month.append(row[0])
        temperature.append(float(row[3]))

plt.plot(month, temperature, label="Temp")
plt.xlabel("Month")
plt.ylabel("°C")
plt.title("Tokyo monthly mean temperature in 2020")
plt.legend()
plt.show()
```

## 発展課題 4

Code セルを追加し 、 以下のようなグラフを描画するプログラムを作成してください

- 月ごとの最高気温・最低気温・平均気温を重ねて表示
- 平均気温の線の種類は 、 plat メソッドの引数 linestyle に指定できます


```python
import matplotlib.pyplot as plt
import csv

month = []
average_temp = []
max_temp = []
min_temp = []
with open("weather_tokyo_2020.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        month.append(row[0])
        max_temp.append(float(row[1]))
        min_temp.append(float(row[2]))
        average_temp.append(float(row[3]))

plt.plot(month, max_temp, label="Highest Temp")
plt.plot(month, min_temp, label="Lowest Temp")
plt.plot(month, average_temp, label="Ave Temp", linestyle="--")
plt.xlabel("Month")
plt.ylabel("°C")
plt.title("Tokyo monthly temperature in 2020")
plt.legend()
plt.show()
```

## 発展課題 5

Code セルを追加し 、 以下のようなグラフを描画するプログラムを作成してください

- 降水量の棒グラフを表示
- 棒グラフの使い方については 、 以下を参考にしてください


```python
import matplotlib.pyplot as plt
import csv

month = []
amount_of_rain = []
with open("weather_tokyo_2020.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        month.append(row[0])
        amount_of_rain.append(float(row[4]))

plt.bar(month, amount_of_rain, label="Rainfall")
plt.xlabel("Month")
plt.ylabel("mm")
plt.title("Tokyo monthly rainfall in 2020")
plt.legend()
plt.show()
```
